```python
import tkinter as tk
from tkinter import messagebox

# Function to check if any player has won and return the winning combination
def check_win(player):
    for i in range(3):
        if all(buttons[i][j]["text"] == player for j in range(3)):
            return [(i, j) for j in range(3)]
        if all(buttons[j][i]["text"] == player for j in range(3)):
            return [(j, i) for j in range(3)]

    if all(buttons[i][i]["text"] == player for i in range(3)):
        return [(i, i) for i in range(3)]
    if all(buttons[i][2 - i]["text"] == player for i in range(3)):
        return [(i, 2 - i) for i in range(3)]

    return None

# Function to check if the board is full (a draw)
def is_board_full():
    return all(buttons[i][j]["text"] != "" for i in range(3) for j in range(3))

# Function to handle button clicks
def on_click(row, col):
    global current_player, game_over

    if buttons[row][col]["text"] == "" and not game_over:
        buttons[row][col]["text"] = current_player

        winning_combination = check_win(current_player)
        if winning_combination:
            for i, j in winning_combination:
                buttons[i][j]["fg"] = "red"  # Change color of winning combination
            messagebox.showinfo("Tic Tac Toe", f"Player {current_player} wins!")
            game_over = True

        elif is_board_full():
            messagebox.showinfo("Tic Tac Toe", "It's a draw!")
            game_over = True

        else:
            # Switch player
            current_player = "X" if current_player == "O" else "O"
            label.config(text=f"Player {current_player}'s turn")

# Function to restart the game
def restart_game():
    global game_over, current_player

    for i in range(3):
        for j in range(3):
            buttons[i][j]["text"] = ""
            buttons[i][j]["fg"] = "black"  # Reset color to black
    game_over = False
    current_player = "X"
    label.config(text="Player X's turn")

# Create the main window
root = tk.Tk()
root.title("Tic Tac Toe")

# Initialize variables
buttons = [[None for _ in range(3)] for _ in range(3)]
current_player = "X"
game_over = False

# Create buttons and add them to the grid
for i in range(3):
    for j in range(3):
        buttons[i][j] = tk.Button(root, text="", font=("Helvetica", 20), width=5, height=2,
                                  command=lambda row=i, col=j: on_click(row, col))
        buttons[i][j].grid(row=i, column=j)

# Create label to display player's turn
label = tk.Label(root, text="Player X's turn", font=("Helvetica", 16))
label.grid(row=3, column=0, columnspan=3)

# Create Restart button
restart_button = tk.Button(root, text="Restart", font=("Helvetica", 14), command=restart_game)
restart_button.grid(row=4, column=0, columnspan=3)

# Start the main event loop
root.mainloop()
```
